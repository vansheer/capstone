```python
import pandas as pd
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('data/all_scripts.csv')
```

```python
cvect = CountVectorizer(stop_words='english', ngram_range=(1, 2), lowercase=False, max_features=500)
```

```python
df_vect = cvect.fit_transform(df['script'])
```

```python
df_vect.toarray
```

```python
cvect.get_feature_names()
```

```python
df = pd.DataFrame(df_vect.toarray(), columns=cvect.get_feature_names())
df.head()
```
